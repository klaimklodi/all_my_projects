## Цены в магазинах

### Задача проекта

В предложенном датасете находится тысяча единиц продукции крупного ритейлера с ценами в зависимости от даты и множеством признаков. Также, у нас есть вспомогательные датасеты, которые могут помочь нам в исследовании. Наша задача спрогнозировать цену каждого товара на выбранном горизонте.

### Этапы проекта

- Загрузка данных и изучение общей информации
- Предобработка данных
- Исследовательский анализ
- Прогноз
- Выводы

### Загрузка данных и изучение общей информации

**Импортируем необходимые библиотеки**

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import ipywidgets as widgets
from sklearn.linear_model import LinearRegression
```

**Загрузим данные и посмотрим на них**

```python
datasets_names = ['train', 'test', 'holidays_events', 'oil', 'stores', 'transactions']
datasets = []

for i in range(len(datasets_names)):
    datasets.append(pd.read_csv(datasets_names[i]+'.csv'))
```

**Выведем датасеты**

```python
data_tabs = widgets.Tab([widgets.Output() for _ in enumerate(datasets)])
for i, df in enumerate(datasets):
    data_tabs.set_title(i, f'Датасет {datasets_names[i]}')
    with data_tabs.children[i]:
        display(df)

data_tabs
```

В тренировочном датасете имеем цены с 1 января 2013 по 15 августа 2017. Предсказать цены необходимо на период с 16 до 31 августа 2017.

### Предобработка данных

Проверим, нуждаются ли какие-либо датасеты в предобработке

`train` и `test`

Пропусков нет, нужно только изменить тип данных столбца *date*

```python
def preprocess(data): #функция для предобработки train и test 
    data.index = data['id']
    data = data.drop(columns=['id'])
    data['date'] = pd.to_datetime(data['date'])
    return data

train = preprocess(datasets[0])
test = preprocess(datasets[1])
```

Здесь порядок, посмотрим следующие датасеты

`holidays_events`

Изменим тип данных столбца *date*

```python
holidays_events = datasets[2]
holidays_events['date'] = pd.to_datetime(holidays_events['date'])
```

`oil`

Изменим тип данных столбца *date* и заполним пропуски

```python
oil = datasets[3]
oil['date'] = pd.to_datetime(oil['date'])
oil['dcoilwtico'].fillna(method='bfill', inplace=True)
```

`stores`

```python
stores = datasets[4]
stores.index = stores['store_nbr']
stores = stores.drop(columns=['store_nbr'])
```

`transactions`

Изменим тип данных столбца *date*

```python
transactions = datasets[5] 
transactions['date'] = pd.to_datetime(transactions['date'])
```

**Предобработка датасетов закончена**, переходим к следующему этапу

### Исследовательский анализ

Для начала посмотрим, как влияли цены на бензин на цены в магазинах

Посчитаем сумму продаж по всем категориям и магазинам на каждую дату и сгладим график, чтобы посмотреть на тренд. Добавим даты праздников

```python
trend = pd.DataFrame(train.groupby('date')['sales'].sum()).rolling(window=31, center=True).mean()
```

Возьмем даты национальных праздников и отразим их на графике красными вертикальными полосами

```python
hol_dates = holidays_events[(holidays_events['date']>='2013-01-01')&(holidays_events['locale']=='National')]['date']
```

```python
fig, ax = plt.subplots(3, 1, figsize=(15,6), sharex=True)
sns.scatterplot(data=trend, x='date', y='sales', ax=ax[0])
ax[0].set_title('Средняя суммы продаж на каждую дату')
ax[0].set_ylabel('Выручка')

for xc in hol_dates:
    ax[0].axvline(x=xc, color='r', alpha=0.1)
    
    
ax[1].plot(transactions.groupby('date')['transactions'].sum().rolling(window=31, center=True).mean())
ax[1].set_title('Суммарное количество покупок во всех магазинах')
ax[1].set_ylabel('Покупок, шт.')

sns.lineplot(data=oil, x='date', y='dcoilwtico', ax=ax[2])
ax[2].set_title('Цена на бензин на каждую дату')
ax[2].set_ylabel('Цена')
ax[2].set_xlabel('Дата')


plt.show()
```

Перед большими скоплениями национальных праздников, всегда наблюдается рост суммарных продаж по всем магазинам, что объясняется резким ростом количества покупок. Период, на который необходимо построить прогноз, не содержит больших кластеров национальных праздников, поэтому можем далее не учитывать их даты в признаках

**Выясним, влияет ли день недели на продажи**

Сначала объединим тренировочный и тестовый датасеты

```python
test['test'] = 1
train['test'] = 0

data = pd.concat([train, test], axis=0)
```

Добавим признаки, показывающие год, месяц, день недели по дате

```python
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['day_of_week'] = data['date'].dt.day_of_week
```

Напишем функцию для построения линейного графика

```python
def plot_line(par, title=None, xlab=None, ylab=None, is_rotation=False):
    fig, ax = plt.subplots(figsize=(10, 5))
    sns.lineplot(data=data[data['test']==0].groupby(par)['sales'].sum().reset_index(), x=par, y='sales', ax=ax)
    
    ax.set_title(title)
    ax.set_xlabel(xlab)
    if is_rotation:
        plt.xticks(rotation=90)
        
    ax.set_ylabel(ylab)

    plt.show()
```

```python
plot_line('day_of_week', 'Суммарные продажи по дням недели', 'День недели', 'Продажи')
```

Видим, что наибольшие продажи приходятся на субботу и воскресенье

**Как отличаются продажи в зависимости от категории товаров?**

```python
plot_line('family', 'Суммарные продажи по категориям товаров', 'Категория', 'Продажи', True)
```

Напитки, бакалея и продукты - категории товаров, которые покупают больше всего

**Отличаются ли продажи в разных магазинах?**

```python
plot_line('store_nbr', 'Суммарные продажи по магазинам', 'Номер магазина', 'Продажи')
```

Оказывается, суммарные продажи заметно отличаются по магазинам сети

### Прогноз

Прежде всего, подготовим признаки, которые будут использоваться в нашей моделе

```python
data = pd.get_dummies(data)
```

Выделим тренировочную и тестовые выборки

```python
train = data[data['test']==0]
test = data[data['test']==1]
target = 'sales'
features = data.drop(columns=['date', 'sales']).columns

X_train, X_test, y_train, y_test = train[features], test[features], \
            train[target], [] 
```

Зададим модель и обучим ее

```python
model = LinearRegression()

model.fit(X_train, y_train)

y_test = pd.Series(model.predict(X_test[features]), index=X_test.index)
```

Сохраним прогнозы в файл

```python
submission = pd.DataFrame(y_test, columns=['sales']).reset_index()
submission.to_csv('my_submission.csv', index=False)
```
