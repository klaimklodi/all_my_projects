Дубликатов нет

**Пропуски в данных**

Посмотрим, сколько пропусков имеем в каждом столбце

**Train data**

```python
train_col_with_missing = [(col, data_train[col].isna().sum())  for col in data_train.columns if data_train[col].isna().sum()>0]
sorted(train_col_with_missing, key=lambda x: x[1])
```

**Test data**

```python
test_col_with_missing = [(col, data_test[col].isna().sum())  for col in data_test.columns if data_test[col].isna().sum()>0]
sorted(test_col_with_missing, key=lambda x: x[1])
```

Видим, что в данных огромное количество пропусков. Займемся их обработкой

По документации, для большинства признаков наличие пропуска означает **отсутствие** этого признака в доме, поэтому попробуем грубо заполнить пропуски для категориальных признаков словом **"No"**, а для количественных **нулями**

```python
def fill_na(data, columns):
    for col in columns:
        if data.dtypes[col[0]] == 'object':
            data[col[0]] = data[col[0]].fillna('No')
        else:
            data[col[0]] = data[col[0]].fillna(0)
    return data
```

```python
data_train = fill_na(data_train, train_col_with_missing)
data_test = fill_na(data_test, test_col_with_missing)
```

```python
data_train.isna().sum().sum(), data_test.isna().sum().sum()
```

Теперь отберем столбцы, в которых 10 и более уникальных значений и посмотрим на распределния

```python
k = 0
cols = [col for col in data_train.select_dtypes(include=['object']).columns if data_train[col].nunique() >= 10]
for i in range(len(cols)):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20, 4))
    try:
        sns.scatterplot(data=data_train, x = cols[k], y = 'SalePrice', hue='SalePrice', ax=ax)
        k+=1      
    except IndexError:
        pass
plt.show()
```

Делаем вывод, что каждый из этих признаков значительно влияет на цену дома, поэтому необходимо оставить их в датасете

**Теперь посмотрим на количественные признаки**

Посчитаем корреляции, чтобы оценить влияние каждого признака на целевую переменную

```python
plt.figure(figsize=(10,7))
plt.title('Корреляции количественных признаков')
sns.heatmap(data_train.select_dtypes(exclude=['object']).corr(), cmap='YlOrRd_r')
plt.show()
```

Видим, что есть есть признаки, которые значительно коррелируют с целевой переменной, отберем их

```python
corr_numeric = data_train.select_dtypes(exclude=['object']).corr()['SalePrice']
numeric_cols = corr_numeric[abs(corr_numeric) > 0.3].index[:-1].tolist()
numeric_cols
```

В результате получили количественные признаки, которые по модулю имеют корреляцию с целевой переменной больше 0.3

**Получим все признаки, которые будут использоваться для предсказания стоимости домов**

```python
good_cols = data_train.select_dtypes(include=['object']).columns.tolist() + numeric_cols + ['SalePrice']
```

```python
data_train = data_train[good_cols]
data_test = data_test[good_cols[:-1]]
data_train.info()
```

### Предсказание стоимости домов

Разделим данные на тестовую и валидационную выборки

```python
y = data_train.SalePrice
data_train.drop(['SalePrice'], axis=1, inplace=True)

X_train, X_valid, y_train, y_valid = train_test_split(data_train, y, test_size=0.2, random_state=0)
```

Будем разбираться с категориальными данными с помощью OrdinalEncoder

```python
obj_cols = [col for col in X_train.columns if X_train[col].dtype == 'object']

ordinal_encoder = OrdinalEncoder()
X_train[obj_cols] = ordinal_encoder.fit_transform(X_train[obj_cols])
X_valid[obj_cols] = ordinal_encoder.fit_transform(X_valid[obj_cols])
```

Зададим модель, обучим ее и вычислим *mean_absolute_error*

```python
model =  XGBRegressor(n_estimators=300, learning_rate=0.05, early_stopping_rounds=2)
model.fit(X_train, y_train, 
             eval_set=[(X_valid, y_valid)], 
             verbose=False)

preds = model.predict(X_valid)

mean_absolute_error(y_valid, preds)
```

**Тестовая выборка**

Получим предсказания по тестовому датасету и згарузим их в csv файл

```python
data_test[obj_cols] = ordinal_encoder.fit_transform(data_test[obj_cols])
```

```python
predictions = model.predict(data_test)
output = pd.DataFrame({'Id': data_test.index, 'SalePrice': predictions})
output.to_csv('my_submission.csv', index=False)
```
