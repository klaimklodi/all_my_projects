**Импортируем библиотеки**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Считаем датасет и посмотрим на него**

```python
data = pd.read_csv('world-data-2023.csv', decimal='.')
data.info()
```

Имеем множество пропусков, проблемы с типами данных. Датасет требует тщательной предобработки данных

### Предобработка данных
<a id='preprocessing'></a>

**Проверим данные на наличие дубликатов**

```python
data['Country'].duplicated().sum()
```

Дубликатов нет!

**Посмотрим на названия столбцов**

```python
data.columns
```

Переименуем столбец *Density\n(P/Km2)*

```python
data = data.rename(columns={'Density\n(P/Km2)' : 'Density (P/Km2)'})
```

Теперь со столбцами порядок

**Посмотрим на пропуски в данных**

Посмотрим, какие столбцы имеют пропуски. Выведем в формате: (столбец, количество пропусков)

```python
na_cols = [(col, data[col].isna().sum())   for col in data.columns if data[col].isna().sum() > 0]
print('Столбцов с пропусками:', len(na_cols))
sorted(na_cols, key=lambda x: x[1], reverse=True)
```

У нас 33 столбца с пропусками. Разберемся с ними

Выясним есть ли страны, которые содержат пропуски больше, чем в половине признаков

```python
data[data.isna().sum(axis=1) > 17]
```

Удалим эти страны из дальнейшего анализа из-за колоссального количества пропусков и посмотрим, сколько осталось пропусков

```python
data = data[data.isna().sum(axis=1) <= 17]

na_cols = [(col, data[col].isna().sum())   for col in data.columns if data[col].isna().sum() > 0]
print('Столбцов с пропусками:', len(na_cols))
sorted(na_cols, key=lambda x: x[1], reverse=True)
```

Выясним, сколько строк имеют пропуски, а сколько нет

```python
print('Строк с пропусками:', len(data[data.isna().sum(axis=1) > 0]))
print('Строк без пропусков:', len(data[data.isna().sum(axis=1) == 0]))
```

Продолжим обработку отстальных пропусков

Заполним средним значением по всем странам пропуски в следующих столбцах:
- *Minimum wage*
- *Tax revenue (%)*
- *Armed Forces size*
- *Gasoline Price*
- *Population: Labor force participation (%)*
- *Unemployment rate*
- *CPI*
- *CPI Change (%)*
- *Maternal mortality ratio*
- *Gross tertiary education enrollment (%)*
- *Total tax rate*
- *Out of pocket health expenditure*
- *Life expectancy*
- *Agricultural Land( %)*
- *Co2-Emissions*
- *Fertility Rate*
- *Forested Area (%)*
- *Gross primary education enrollment (%)*
- *Physicians per thousand*
- *Birth Rate*
- *Infant mortality*<br>

Удалим столбцы, которые не будут рассматриваться далее:
- *Abbreviation*
- *Largest city*
- *Capital/Major City*
- *Official language*
- *Latitude*
- *Longitude*
- *Currency-Code*




```python
data = data.drop(columns=['Abbreviation', 'Largest city', 'Capital/Major City', 'Official language', 'Latitude', 'Longitude',
                         'Currency-Code'])
```

Перед тем, как обрабатывать пропуски, изменим **типы данных**

```python
for col in data.columns[1:]:
    data[col] = [str(x).replace('$', '') for x in data[col]]
    data[col] = [str(x).replace(',', '') for x in data[col]]
    data[col] = [str(x).replace('%', '') for x in data[col]]
    
data[data.columns[1:]] = data[data.columns[1:]].astype(float)
```

```python
data.info()
```

Теперь с типами данных порядок! Продолжим обработку пропусков

```python
na_cols = [col for col in data.columns if data[col].isna().sum() > 0]

for col in na_cols:
    data[col].fillna(data[col].mean(), inplace=True)
```

```python
data.info()
```

Пропуски заполнены! Можно переходить к исследовательскому анализу

### Исследовательский анализ
<a id='analysis'></a>

**Посмотрим на самые густонаселенные страны**

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(data=data.sort_values('Population', ascending=False).head(15), x='Country', y='Population', ax=ax)
ax.set_title('Самые густонаселенные страны')
ax.set_xlabel('Страны')
ax.set_ylabel('Население, млрд')
plt.xticks(rotation=45)
plt.show()
```

Не новость, что Китай и Индия - лидеры по населению. <br>
**Посмотрим, как коррелирует численность населения с плотностью**

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.scatterplot(data=data[(data['Population']<1*10**8)&(data['Density (P/Km2)']<1000)], x='Density (P/Km2)', y='Population', ax=ax)
ax.set_title('Корреляция численности и плотности населения')
ax.set_xlabel('Плотность населения на кв. км')
ax.set_ylabel('Население')
plt.xticks(rotation=45)
plt.show()
```

Видим, что плотность населения, в целом, не коррелирует с численностью населения

**Посмотрим, как распределена численность вооруженных сил по странам**

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
sns.boxplot(data=data, y='Armed Forces size', ax=ax[0])
ax[0].set_title('Распределение численности вооруженных сил', fontsize=10)
ax[0].set_ylabel('Численность вооруженных сил, млн')

sns.boxplot(data=data[data['Armed Forces size']<5*10**5], y='Armed Forces size', ax=ax[1])
ax[1].set_title('Распределение численности вооруженных сил (масштаб)', fontsize=10)
ax[1].set_ylabel('Численность вооруженных сил')

plt.show()
```

У некоторых стран численность армии доходит до 3 млн. человек. 75% стран имеют численность вооруженных сил до 150 000 человек

**Посмотрим, растет ли процент сельскохоз-й земли с увеличением площади стран**

```python
sns.lmplot(data=data, x='Land Area(Km2)', y='Agricultural Land( %)', ci=None, height=6, aspect=2)
plt.show()
```

По графику получается, что с ростом площади страны незначительно снижается размер ее сельскозяйственных территорий. Грубо говоря, можем сделать вывод, что показатели не коррелируют

**Посмотрим, как респределен уровень рождаемости по миру**

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.kdeplot(data=data, x='Birth Rate', multiple="stack", ax=ax)
ax.set_title('Распределение уроня рождаемости по миру')
ax.set_ylabel('Частота')


plt.show()
```

*Birth Rate* показывает, сколько рождается детей каждый год на 1000 населения страны. Видим, что по миру показатель чаще всего принимает значение около 10

**Выясним, влияет ли рождаемость на ожидаемую продолжительность жизни**

```python
sns.jointplot(data=data, x='Birth Rate', y='Life expectancy')

plt.show()
```

Наглядно видим отрицательную корреляцию: Чем выше уровень рождаемости, тем меньше ожидаемая продолжительность жизни

**Посмотрим, на распределение цен на бензин по миру**

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.violinplot(data=data, y='Gasoline Price', ax=ax)
ax.set_title('Распределение цены на литр бензина')
ax.set_ylabel('Цена бензина за литр в местной валюте')

plt.show()
```

Литр бензина в среднем по миру стоит 1 доллар

**Посмотрим, в каких странах выбрасывается больше всего углекислого газа**

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(data=data.sort_values('Co2-Emissions', ascending=False).head(15), x='Country', y='Co2-Emissions', ax=ax)
ax.set_title('Выбросы углекислого газа по странам')
ax.set_xlabel('Страны')
ax.set_ylabel('Выбросы газа, в тоннах')
plt.xticks(rotation=45)
plt.show()
```

Ничего неожиданного, Китай, США и Индия загрязняют нашу планету больше всего.<br> **Проверим, коррелируют ли ВВП и выбросы углекислого газа**

```python
sns.lmplot(data=data[data['Co2-Emissions']<1000000], x='Co2-Emissions', y='GDP', ci=None, height=6, aspect=2)
plt.show()
```

Радует, что большое количество выбросов углекислого газа объясняется ростом ВВП

**Сравним размеры налогов по странам**

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(data=data.sort_values('Tax revenue (%)', ascending=False).head(15), y='Country', x='Tax revenue (%)', ax=ax, orient='h')
ax.set_title('Размер налогов по странам')
ax.set_ylabel('Страны')
ax.set_xlabel('Размер налоговых поступлений от ВВП, в процентах')
plt.xticks(rotation=45)
plt.show()
```

В некоторых странах размер налоговых поступлений составляет более 30% от ВВП. Среди них: Алжир, Сейшельские Острова, Дания и Лесото. <br> **Узнаем, влияет ли размер налога на ВВП страны**

```python
sns.lmplot(data=data, x='Tax revenue (%)', y='GDP', ci=None, height=6, aspect=2)
plt.show()
```

Видимо деньги, получаемые с населения идут не на рост ВВП. Видим отрицательную корреляцию
